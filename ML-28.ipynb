###### (2)Random Over Sampling - We will increase the size of minority is inactive class too the size of majority class is active.

##### Under Sampling

```python
from imblearn.under_sampling import RandomUnderSampler
```

```python
ru = RandomUnderSampler()
ru.fit_resample(x, y)
```

```python
ru_x,ru_y = ru.fit_resample(x, y)
```

```python
ru_x
```

```python
ru_y
```

```python
# lets check the large data 0 is now reduced or not
ru_y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(ru_x,ru_y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr =LogisticRegression()
lr.fit(x_train, y_train)
```

```python
lr.score(x_test, y_test)
```

```python
# Here we can see that the model is performing well because of the balanced dataset
lr.predict([[25	,40150]])
```
