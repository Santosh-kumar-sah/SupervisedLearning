### SVM Concepts

##### Margin:
##### - Hard Margin: Used when the data is linearly separable. It does not tolerate any misclassification. The goal is to find the hyperplane that maximizes the margin with no points inside the margin.
##### - Soft Margin: Used when the data is not perfectly linearly separable. It allows for some misclassifications by introducing a slack variable. It balances maximizing the margin and minimizing the number of misclassified points.

##### Types of SVM:
##### 1.  **Linear SVM**: For data that can be separated by a single straight line (hyperplane).
##### 2.  **Non-linear SVM**: For data that is not linearly separable. It uses the kernel trick to map data to a higher dimension where a linear separator can be found.

##### **Kernel Functions**:
##### Kernels are functions that take low-dimensional input space and transform it into a higher-dimensional space.

##### 1.  Linear Kernel:
#####    -   `K(x, y) = x^T * y`
##### 2.  Polynomial Kernel:
#####    -   `K(x, y) = (gamma * x^T * y + r)^d`
##### 3.  Radial Basis Function (RBF) Kernel:
#####    -   `K(x, y) = exp(-gamma * ||x - y||^2)`
##### 4.  Sigmoid Kernel:
#####    -   `K(x, y) = tanh(gamma * x^T * y + r)`

##### Classification

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
dataset = pd.read_csv("student_data_binary.csv")
```

```python
dataset.head(3)
```

```python
sns.scatterplot(data=dataset, x="CGPA", y="Score", hue="Placed")
plt.show()
```

```python
x = dataset.iloc[:, :-1]
y = dataset["Placed"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=9)
```

```python
from sklearn.svm import SVC
```

```python
# kernel = linear, rbf, poly,sigmoid,precomputed
sv = SVC(kernel="linear")
sv.fit(x_train, y_train)
```

```python
sv.score(x_test, y_test)*100
```

```python
sv.score(x_train, y_train)*100
```

```python
plot_decision_regions(np.array(x), np.array(y), clf=sv, legend=2)
plt.show()
```
