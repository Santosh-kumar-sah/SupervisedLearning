### Multiple Linear Regression-- Having multiple features

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("employee.csv")
```

```python
dataset.head(3)
```

```python
dataset.shape
```

```python
dataset.isnull().sum()
```

```python
sns.pairplot(dataset)
plt.show()
```

```python
sns.heatmap(data=dataset.corr(), annot=True)
plt.show()
```

```python
x = dataset.iloc[:, :-1]
y = dataset["Salary"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# dimensions of x
x.ndim
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
# fit will show this y = m1x1 + m2x2 + m3x3 + ... + c
lr.fit(x_train, y_train)
```

```python
lr.score(x_test, y_test)*100
```

```python
# coefficients 
lr.coef_
```

```python
# intercept_
lr.intercept_
```

```python
# our Equation is
# y = m1x1 + m2x2 + m3x3 + ... + c
# y = 1663.5122*Age + 103.1946*Experience
```
