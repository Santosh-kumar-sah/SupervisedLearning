### Label Encoding

Categorical Data-:

(1)NOMINAL DATA - The data which does not have any connections between them.
(2)Ordinal Data-It has some sequence order

```python
import pandas as pd
```

```python
df = pd.DataFrame({"Name":["Ram","Shyam","Madan","Mohan","Sundar"]})
df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
# We do fit_transform at same time while training and while deploying we do fit and then transform
le.fit_transform(df["Name"])
```

```python
# We have add this in the given dataset
df["en_Name"] =le.fit_transform(df["Name"])
df
```

```python
# Lets work with real data
dataset = pd.read_csv("loan.csv")
dataset.head(5)
```

```python
# Lets check what type of data is present
dataset["Property_Area"].unique()
```

```python
# using label Encoder
la = LabelEncoder()
la.fit(dataset["Property_Area"])
la.transform(dataset["Property_Area"])
```

```python
# Replace the existing data like urban, rural with the numbers then(see Result in 2nd cell above)
dataset["Property_Area"] = la.transform(dataset["Property_Area"])
```
