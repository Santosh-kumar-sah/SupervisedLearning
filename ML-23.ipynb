##### it can be  or 1, True or False etc. But instead of giving the exact value 0 and 1 gives probabilistic values between 0 and 1

```python
import pandas as pd
import numpy as np      
import matplotlib.pyplot as plt
import seaborn as sns
```

#### It has only 1 Feature (Single Feature)

```python
dataset = pd.read_csv('social_network_ads_single.csv')
dataset.head(3)
```

```python
sns.scatterplot(data=dataset, x="Age", y="Purchased" )
plt.show()
```

```python
x = dataset[["Age"]]
y = dataset["Purchased"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression


```

```python
lr = LogisticRegression()
lr.fit(x_train, y_train)
```

```python
lr.score(x_test, y_test)
```

```python
y_pred = lr.predict(x)
```

```python
sns.scatterplot(data=dataset, x="Age", y="Purchased" )
sns.lineplot(x= "Age",y =lr.predict(x), data=dataset,color='red')
plt.show()
```
