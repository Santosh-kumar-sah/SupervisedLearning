### Outlier and Handling it

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset =pd.read_csv("loan.csv")
dataset.info()
```

```python
dataset.describe()
```

```python
# using BoxPlot we can find outliers
sns.boxplot(x="ApplicantIncome",data=dataset)
plt.show()
```

```python
# We can also see by using distribution (a bit complex to observe)
sns.histplot(dataset["ApplicantIncome"], bins=30, kde=True)
plt.show()
```
