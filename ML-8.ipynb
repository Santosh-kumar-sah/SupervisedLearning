### Removing outlier using IQR method(Inter quantile Range)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("loan_dataset.csv")
dataset.head(5)
```

```python
dataset.describe()
```

```python
plt.figure(figsize =(10,7))
sns.boxplot(x="ApplicationIncome",data=dataset)
plt.show()
```

#### IQR = Q3 - Q1
##### first min range = Q1 - (1.5 * IQR)
##### max range = Q3 + (1.5* IQR)

```python
dataset.shape
```

```python
q1 = dataset["ApplicationIncome"].quantile(0.25)

q3 = dataset["ApplicationIncome"].quantile(0.75)
```

```python
IQR = q3 - q1
```

```python
min_range = q1 - (1.5 * IQR)
max_range = q3 + (1.5 * IQR)

min_range,max_range
```

```python
# filtering data less than max_Range

new_dataset = dataset[dataset["ApplicationIncome"]<=max_range]
```

```python
new_dataset.shape
```

```python
# plotting for new Dataset
sns.boxplot(x="ApplicationIncome",data=new_dataset)
plt.show()
```
