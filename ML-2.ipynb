### Handling missing value(imputing categorical Data)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset= pd.read_csv("Data.csv")
```

```python
# filling data(not right way to fill)
dataset.fillna(10)
```

```python
# right way(it gives some warning)
# dataset.fillna(method ="bfill")
# axis =1 -->col and axis =0 -->row
dataset.ffill(axis=1)
```

```python
#  one way to fill categorical data is by Mode(finding Mode)
dataset["Name"].mode()
```

```python
dataset["Name"].mode()[0]
```

```python
# filling the dataset Name by "Alice"
dataset["Name"].fillna(dataset["Name"].mode()[0],inplace=True)
```

```python
# collecting object dataType
dataset.select_dtypes(include="object")
```

```python
# finding null value
dataset.select_dtypes(include="object").isnull().sum()
```

```python
# using loop fining categorical data
for i in dataset.select_dtypes(include="object").columns:
    print(i)
```

```python
for col in dataset.select_dtypes(include="object").columns:
    dataset.loc[:, col] = dataset[col].fillna(dataset[col].mode()[0])

    
        
```

```python
dataset.isnull().sum()
# it shows that categorical data is filled by mode
```
