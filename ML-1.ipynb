```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset= pd.read_csv("Data.csv")

dataset.head(10)
```

```python
dataset.shape
```

```python
dataset.isnull()
```

```python
# gives the col in which there is missing value
dataset.isnull().sum()
```

```python
dataset.shape[0]
```

```python
# total number of null value
dataset.isnull().sum().sum()
```

```python
# percantage of null value(here shape[0]==8 that is rows) in each
(dataset.isnull().sum()/dataset.shape[0])*100
```

```python
# overall null value percentage
(dataset.isnull().sum().sum()/(dataset.shape[0]*dataset.shape[1]))*100
```

```python
# Gives the filled data
dataset.notnull().sum()
```

```python
# Gives the TOTAL filled data
dataset.notnull().sum().sum()
```

```python
sns.heatmap(dataset.isnull())
plt.show()
```
