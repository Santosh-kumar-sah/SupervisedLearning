#### Cross-Validation - techinque for validating the model efficiency by training it on the subset of input data and testing on previously unseen subset of input data.

##### Methods used for Cross-Validation-
###### (1)Leave p out cross-validation
###### (2)Leave one out cross-validation
###### (3)Holdout cross validation
###### (4)Repeated random subsampling validation
###### (5) k-fold cross-validation
###### (6) Stratified k-fold cross-validation
###### (7)Time Series cross-validation
###### (8)Nested cross validation

#### (1) K-Fold Cross Validation
##### in this dataset is equally partitioned into k subparts or folds. Out of the k-folds or groups for each iteration one group is selected as validation data and the remaining (k-1) groups are selected as training data. (Not suitable for an imbalanced dataset).

#### (2) Stratified K fold CV(Imbalanced data)(sabhi tarah ke classification hona chahiye)
##### in this dataset is equally partitioned into k subparts or folds. Out of the k-folds or groups for each iteration one group is selected as validation data and the remaining (k-1) groups are selected as training data

##### (3)Leave one out - ek data ko training me baaki rest ko testing me (It is slow for large dataset)

##### (4)Leave P out Cross validation - Ek se jayda data ko trainnig me rest ko testing me dalta h (p as validation data and remaining training data)(it is also slow)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("cgpa_vs_package.csv")
dataset.head(3)
```

```python
x = dataset.iloc[:, :-1]
y = dataset["Package(LPA)"]
```

```python
new_data = dataset.head(10)
new_data
```

```python
x_new = new_data.iloc[:, :-1]
y_new = new_data["Package(LPA)"]
```

```python
from sklearn.model_selection import LeaveOneOut,LeavePOut,KFold,StratifiedKFold
```

```python
# Leave One Out Cross Validation
lo = LeaveOneOut()

for train, test in lo.split(x_new,y_new):
    print("Train Index: ",train,"Test Index: ",test)
```

```python
lp = LeavePOut(p=2)

for train, test in lp.split(x_new,y_new):
    print("Train Index: ",train,"Test Index: ",test)
```

```python
# nsplit = 5 means kitne part me data jayega test ke liye aur kitne part me training ke liye
kf = KFold(n_splits=5)

for train, test in kf.split(x_new):
    print("Train Index: ",train,"Test Index: ",test)
```

```python
# works only with classification problems(So it will not work here)
skf = StratifiedKFold(n_splits=5)

# for train, test in skf.split(x_new,y_new):
#     print("Train Index: ",train,"Test Index: ",test)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# Finding the range of accuracy
from sklearn.model_selection import cross_val_score
```

```python
# cross_val_score(estimator,features,target,cv=number of splits)
# Or cv = kf or lo or lp or skf (jo bhi use karna ho)
cross_val_score(LinearRegression(),x,y,cv=5)  # 5 means 5 fold cross validation
```

```python
# Arranging in sorted manner
p = cross_val_score(LinearRegression(),x,y,cv=5)
p.sort()
p
```

```python
# if use cv = kf or lo or lp or skf
p = cross_val_score(LinearRegression(),x,y,cv=KFold(n_splits=10))
p.sort()
p
```
