#### Logistic regression(Binary)(Multiple Features)
##### contains more than 1 feature

```python
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('placement_data.csv')
dataset.head(3)
```

```python
# to check linearly separable or not we use "hue" parameter
sns.scatterplot(data=dataset, x='CGPA', y='Score', hue='Placed')
plt.show()
```

```python
# We take x 2D and iloc gives 2D array
x = dataset.iloc[:, :-1]

y = dataset["Placed"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr.fit(x_train, y_train)
```

```python
lr.score(x_test, y_test)
```

```python
# just checking the prediction by taking some random values
lr.predict([[6.87,419]])
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
# This function requires numpy arrays
plot_decision_regions(x.to_numpy(), y.to_numpy(), clf=lr)
plt.xlabel('CGPA')
plt.ylabel('Score') 
plt.title('Logistic Regression Decision Boundary')
plt.show()
```
