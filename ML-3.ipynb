# Handling missing values with Scikit Learn


```python
import pandas as pd
```

```python
dataset = pd.read_csv("Data.csv")
dataset.head(10)
```

```python
dataset.info()
```

```python
dataset.select_dtypes(include="float64").columns
```

```python
from sklearn.impute import SimpleImputer
```

```python
# We are filling the Missing values by using Mean
si = SimpleImputer(strategy="mean")
si.fit_transform(dataset[['Age', 'Salary', 'PerformanceScore']])
```

```python
# arranging in a dataframe
# storing the si value in a variable
ar=si.fit_transform(dataset[['Age', 'Salary', 'PerformanceScore']])
pd.DataFrame(ar,columns=dataset.select_dtypes(include="float64").columns)
```

```python
# Now checking the missing value present or not
new_dataset=pd.DataFrame(ar,columns=dataset.select_dtypes(include="float64").columns)
new_dataset.isnull().sum()
```

```python
# Here we can see the filled missing value
new_dataset
```
