### One Hot Encoding And Dummy Variables

process of converting categorical data(like string, object etc) into numerical data is called Encoding

Basically it is used when there is like YES/NO or Male/Female type data


```python
import pandas as pd
```

```python
dataset = pd.read_csv("Large_Employee_Data.csv")
dataset.head(10)
```

```python
dataset.isnull().sum()
```

```python
# filling dataset(This method may give warning so use below cell method)
dataset["Gender"].fillna(dataset["Gender"].mode()[0],inplace=True)
```

```python

# This is the new method used in new Version of python
dataset["Marital Status"] = dataset["Marital Status"].fillna(dataset["Marital Status"].mode()[0])
```

```python
# Before we apply Get dummies we have to select the columns 
en_data = dataset[["Gender","Marital Status"]]
en_data
```

```python
# One hot Encoding using get Dummy Method
pd.get_dummies(en_data)
```

```python
# Checking in data info
pd.get_dummies(en_data).info()
```

```python
# From skLearn we are doing one Hot Encoding
from sklearn.preprocessing import OneHotEncoder
```

```python
# Using one hot encoding (it gives a sparse Matrix)
OnehotEn = OneHotEncoder()
OnehotEn.fit_transform(en_data)
```

```python
# Converting into array
arrayData=OnehotEn.fit_transform(en_data).toarray()
```

```python
pd.DataFrame(arrayData,columns=["Gender_Female","Gender_Male","Marital Status_Married","Marital Status_Single"])
```
