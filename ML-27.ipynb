#### Confusion Matrix -understanding the performance of model, how well model is doing, number of correct predictions and incorrect predictions

##### In this we are studying according Binary Class classification later on Multiclass classification

##### Accuracy = TP+TN/n

##### Error = FN+FP/n
#####
#####

##### False Negative - The model has predicted no, but the actual value was Yes, it is also called as Type-II error
##### False Positive - The model has predicted Yes, but the actual value was No, it is also called as Type-I error

##### Precision - Measure the ability to classify psositive samples in the model
##### Precision = TP/TP+FP 

##### Recall- Measure how many positive samples were correctly classified by the ML Model
##### Recall = TP/TP+FN

##### F1-Score - Harmonic mean of precision and recall.It takes both false positive and false negative into account.Performs well on an imbalanced dataset.

##### F1 Score = 2 * P *R/P+R

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("placement_data.csv")
dataset.head(3)
```

```python
x = dataset.iloc[:, :-1]
y = dataset["Placed"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr.fit(x_train, y_train)
```

```python
lr.score(x_test, y_test)
```

```python
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score
```

```python
confusion_matrix(y_test, lr.predict(x_test))
```

```python
cf =confusion_matrix(y_test, lr.predict(x_test))
```

```python
sns.heatmap(cf, annot=True)
plt.show()
```

```python
precision_score(y_test, lr.predict(x_test))
```

```python
recall_score(y_test, lr.predict(x_test))
```

```python
f1_score(y_test, lr.predict(x_test))
```
