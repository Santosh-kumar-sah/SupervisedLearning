##### SVM(Regression)- for modeling and predicting continuous outcomes.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
dataset = pd.read_csv("cgpa_vs_package.csv")
dataset.head(3)
```

```python
sns.scatterplot(data=dataset, x="CGPA", y="Package(LPA)")
plt.show()
```

```python
x = dataset[["CGPA", "Package(LPA)"]]
y = dataset["Package(LPA)"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.svm import SVR
```

```python
sr = SVR(kernel='linear',degree=3)
sr.fit(x_train, y_train)
```

```python
sr.score(x_test, y_test)*100
```

```python
sr.score(x_train, y_train)*100
```

```python
sns.scatterplot(data=dataset, x="CGPA", y="Package(LPA)")
plt.plot(dataset["CGPA"],sr.predict(x), color='red')
plt.show()
```
