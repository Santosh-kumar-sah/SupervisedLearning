#### Regression Analysis is a statistical method used to model the relationship between a dependent variable and one or more independent variables. The primary goal is to predict the value of the dependent variable.
#### Common Types of Regression:

##### Linear Regression: Models the relationship using a straight line.
    #### Simple Linear Regression: One independent variable.
   Multiple Linear Regression: More than one independent variable.
##### Polynomial Regression: Models a non-linear relationship using a curved line.
##### Logistic Regression: Used for classification tasks to predict a categorical outcome (e.g., Yes/No).
##### Ridge & Lasso Regression:Variations of linear regression that are used to prevent overfitting, especially when dealing with many independent variables.

##### Linear Regression Algorithm(Simple)--> Contains only one feature(1 col)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
x =dataset[["CGPA"]]
y = dataset["Package(LPA)"]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=72) 
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# training the model
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
# find coefficient 
lr.coef_
```

```python
# find intercept
lr.intercept_
```

#### basically the fit function do the calculatino of the equation of the line

```python
# Accuracy of the model
lr.score(x_test,y_test)
# In percentage
lr.score(x_test,y_test)*100
```

##### This is very poor accuracy
##### How to increase accuracy--> By changing the random_state value and note the accuracy which one will largest will be the best Accuracy

```python
# testing the model
lr.predict([[6.87]])
```

```python
y_pred = lr.predict(x)
```

```python
sns.scatterplot(x ="CGPA",y = "Package(LPA)",data = dataset,label="Actual Data")
plt.plot(x_test,y_pred,color="red",label="Predicted line")
plt.show()
```
